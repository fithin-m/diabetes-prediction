```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
# 0---->non diabetic
#1----->diabetic
```

```python
df.groupby('Outcome').mean()
```

```python
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
print(x)

```

```python
print(y)
```

```python
# DATA STANDARDIZATION
scaler=StandardScaler()
standardized_df=scaler.fit_transform(x)
```

```python
print(standardized_df)
```

```python
x=standardized_df
y=df['Outcome']
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the model 1
classifier=svm.SVC(kernel='linear')
```

```python
#training the svc
classifier.fit(x_train,y_train)
```

```python
#ACCURACY SCORE
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy of traing data :',training_data_accuracy)
```

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy of test data :',test_data_accuracy)
```

```python
#training the model 2
classifier2=svm.SVC(kernel='poly')
```

```python
classifier2.fit(x_train,y_train)
```

```python
x_train_prediction=classifier2.predict(x_train)
training_data_accuracyy=accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy of traing data :',training_data_accuracyy)
```

```python
x_test_predictions=classifier2.predict(x_test)
test_data_accuracyy=accuracy_score(x_test_predictions,y_test)
```

```python
print('accuracy of test data :',test_data_accuracyy)
```
